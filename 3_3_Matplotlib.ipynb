Matplotib:

Useful for Making Plots

```python
# importing matplotlib library
import matplotlib.pyplot as plt
```

```python
#importing numpy library to get data for using plots
import numpy as np

x = np.linspace(0, 5, 11)
y =np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

Ploting a Data

```python
#sine wave
plt.plot(x,y)
plt.show()
```

```python
# cosine wave
plt.plot(x,z)
plt.show()
```

```python
# adding title x-axis and y-axis labels
plt.plot(x,y)
plt.title('Sine Wave')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

```python
# plot a parabola

x = np.linspace(-10, 10, 20)
y = x**2    # used for barabola
plt.plot(x,y)
plt.title('Parabola')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

```python
plt.plot(x,y, 'g+')
plt.show()
```

```python
x = np.linspace(-5, 5, 50)
plt.plot(x,np.sin(x), 'r-')
plt.plot(x,np.cos(x), 'b--')
plt.show()
```

BAR Plots

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['Python', 'C++', 'Ruby', 'Java',]
students = [23,17,35,29]
ax.bar(languages, students)
plt.show()
```

Pie chart

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['Python', 'C++', 'Ruby', 'Java',]
students = [23,17,35,29]
ax.pie(students,labels=languages,autopct='%1.1f%%')
plt.show()
```

#Scatter Plot

```python
x = np.linspace(0, 10, 40)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y, color='g')
ax.scatter(x,z,color='r')
plt.show()
```

#3D  scatter Plot

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Greens')
plt.show()
```
